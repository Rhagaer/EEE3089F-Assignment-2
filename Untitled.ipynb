```python
import matplotlib.pyplot as plt
import numpy as np
from sympy import *


inputPower = 40 #[W]
efficency = 1 #Percentage from[0-1]
linearGain = 3.981

init_printing(use_unicode=True)

r = symbols('r')
powerDensity = simplify((inputPower*efficency*linearGain)/(4*np.pi*np.power(r,2)))
powerDensity
```

```python
powerDensityLamda = lambdify(r, powerDensity)
rValues = np.arange(1, 1000, 1)
powerDensityValues = powerDensityLamda(rValues)
dBmValues = 10*np.log10(powerDensityValues)


plt.semilogx(rValues, dBmValues)
plt.title("Power Density Versus Distance from Tower")
plt.xlabel('Distance [m]')
plt.ylabel('dBm')
plt.grid(which='both', axis='both')
plt.show()
```
